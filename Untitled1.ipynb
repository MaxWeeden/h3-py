```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)

n = 5
a = np.random.rand(n) + .1
b = 1
p = 7

x = b/a
x.sort()
x = np.insert(x, 0, 0)

y = np.sum(np.maximum(a[:,np.newaxis]*x, b), axis=0)

i = np.searchsorted(y, p)
f = (p - y[i-1])/(y[i] - y[i-1])
x_star = (1-f)*x[i-1] + f*x[i]

plt.plot(x,y, '-o')
plt.axhline(p, c='r', alpha=.3)
plt.axvline(x_star, c='r', alpha=.3)
```
